```python
import tensorflow as tf
import matplotlib.pyplot as plt
from models import WindowObjectDetector
from matplotlib.patches import Rectangle
```

```python
model = WindowObjectDetector(tf.keras.models.load_model('../checkpoints/resnet50-best.h5'))
x = tf.keras.applications.resnet50.preprocess_input(
    tf.keras.preprocessing.image.img_to_array(
        tf.keras.preprocessing.image.load_img('/tmp/test.png')
    )
)
x.shape = (1,) + x.shape
brow, bcol, boxes = model.predict(x)
# boxes
```

```python
# rowmin, colmin, rowmax, colmax
```

```python
im = tf.keras.preprocessing.image.load_img('/tmp/test.png')
```

```python
chors = boxes[1]

fig, ax = plt.subplots(figsize=(16, 12))
ax.imshow(im)
org_width, org_height = im.size
for rowmin, colmin, rowmax, colmax in chors:
    x = colmin / bcol * org_width
    y = rowmin / brow * org_height
    width = (colmax - colmin) / bcol * org_width
    height = (rowmax - rowmin) / brow * org_height
    ax.add_patch(
        Rectangle((x, y), width, height, fc='r', alpha=0.5)
    )
```
